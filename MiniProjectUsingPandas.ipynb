```python
import  pandas as pd
job_market=pd.read_csv("/content/indian-job-market-dataset-2025.csv")


job_market
```

```python
#total job postings
job_market.value_counts()
```

```python
#no.of unique companies that are hiring
job_market['companyName'].nunique()
```

```python
#no.ofunique job titles
job_market['title'].nunique()
```

```python
#columns contain missing values, and count
job_market.isnull().sum()
```

```python
#duplicate jobId values
job_market['jobId'].duplicated().sum()
```

```python
#percentage of jobs that have salary information missing or undisclosed
(job_market['salary'] == 0).sum() / len(job_market) * 100
```

```python
#What are the top 10 companies with the most job postings?
job_market['companyName'].value_counts().head(10)
```

```python
#What are the top 10 most common job titles?
job_market['title'].value_counts().head(10)
```

```python
#Which 10 locations have the most job opportunities?
job_market['location'].value_counts().head(10)
```

```python
#What is the average minimum and maximum salary across the dataset?
avg_min=job_market['minimumSalary'].mean()
avg_max=job_market['maximumSalary'].mean()
print(int(avg_min))
print(int(avg_max))
```

```python
#Which job posting has the highest maximumSalary
highest_salary_job_index = job_market['maximumSalary'].idxmax()
highest_salary_job_title = job_market.loc[highest_salary_job_index, 'title']
print(f"The job posting with the highest maximum salary is: {highest_salary_job_title}")
```

```python
#Which companies offer the highest average maximum salary?
job_market.groupby('companyName')['maximumSalary'].mean().sort_values(ascending=False).head(1)
```

```python
#Which locations have the highest average salaries?
highest_avg_sal=job_market.groupby('location')['maximumSalary'].mean().sort_values(ascending=False).head(10)
highest_avg_sal
```

```python
#What are the top 10 highest-paying jobs requiring 0–1 years of minimum experience?
job_market[job_market['minimumExperience'] == 0].sort_values(by='maximumSalary', ascending=False).head(10)
```

```python
#How many jobs require zero minimum experience?
zero_min_experience_jobs = job_market[job_market['minimumExperience'] == 0]
num_zero_min_experience_jobs = len(zero_min_experience_jobs)
num_zero_min_experience_jobs
```

```python
#Find the top 20 most demanded skills in the Indian job market from tagsAndSkills
job_market['tagsAndSkills'].str.split(', ').explode().value_counts().head(20)
```
