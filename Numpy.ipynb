marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])

Find each student’s average, each subject’s average, the student with the highest total marks, and all students whose average is above 80.

```python
marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])
stu1=marks[0].mean()
stu2=marks[1].mean()
stu3=marks[2].mean()
stu4=marks[3].mean()
print("Student 1 subjects avg is ",stu1)
print("Student 2 subjects avg is ",stu2)
print("Student 3 subjects avg is ",stu3)
print("Student 4 subjects avg is ",stu4)

sub1=marks[:,0].mean()
sub2=marks[:,1].mean()
sub3=marks[:,2].mean()

print("Subject 1 avg is ",sub1)
print("Subject 2 avg is ",sub2)
print("Subject 3 avg is ",sub3)

total1=marks[0].sum()
total2=marks[1].sum()
total3=marks[2].sum()
total4=marks[3].sum()

print("Total marks of student 1 is ",total1)
print("Total marks of student 2 is ",total2)
print("Total marks of student 3 is ",total3)
print("Total marks of student 4 is ",total4)


```

```python
student_averages = marks.mean(axis=1)
print("Student averages:", student_averages)

students_above_80_avg = np.where(student_averages > 80)[0]

if len(students_above_80_avg) > 0:
    print("Students with average marks above 80:")
    for student_index in students_above_80_avg:
        print(f"  Student {student_index + 1} (Average: {student_averages[student_index]:.2f})")
else:
    print("No students found with average marks above 80.")
```

Normalize this array to the range 0–1:
arr = np.array([10, 20, 30, 40, 50])

Use the min-max normalization formula. Don’t hardcode the minimum or maximum.

```python
arr = np.array([10, 20, 30, 40, 50])
min_val=arr.min()
max_value=arr.max()
normalization = (arr-min_val)/(max_value-min_val)
print("Normalized array ", normalization)
```


sales = np.array([

Find total sales per row, total sales per column, the row with the highest total sales, and replace values below the overall mean with 0.

```python
sales = np.array([
    [100, 200, 150],
    [300, 250, 400],
    [120, 180, 160],
    [500, 450, 550]
])



row_sum1=sales[0].sum()
row_sum2=sales[1].sum()
row_sum3=sales[2].sum()
row_sum4=sales[3].sum()

col_total1=sales[:,0].sum()
col_total2=sales[:,1].sum()
col_total3=sales[:,2].sum()

print("Total sales for row 1 is ",row_sum1)
print("Total sales for row 2 is ",row_sum2)
print("Total sales for row 3 is ",row_sum3)
print("Total sales for row 4 is ",row_sum4)

print("Total sales for column 1 is ",col_total1)
print("Total sales for column 2 is ",col_total2)
print("Total sales for column 3 is ",col_total3)

overall_mean=sales.mean()
print(overall_mean)

sales[sales<overall_mean]=0
print("midified sales data ",sales)

```
