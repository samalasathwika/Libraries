marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])

Find each student’s average, each subject’s average, the student with the highest total marks, and all students whose average is above 80.

```python
marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])
stu1=marks[0].mean()
stu2=marks[1].mean()
stu3=marks[2].mean()
stu4=marks[3].mean()
print("Student 1 subjects avg is ",stu1)
print("Student 2 subjects avg is ",stu2)
print("Student 3 subjects avg is ",stu3)
print("Student 4 subjects avg is ",stu4)

sub1=marks[:,0].mean()
sub2=marks[:,1].mean()
sub3=marks[:,2].mean()

print("Subject 1 avg is ",sub1)
print("Subject 2 avg is ",sub2)
print("Subject 3 avg is ",sub3)

total1=marks[0].sum()
total2=marks[1].sum()
total3=marks[2].sum()
total4=marks[3].sum()

print("Total marks of student 1 is ",total1)
print("Total marks of student 2 is ",total2)
print("Total marks of student 3 is ",total3)
print("Total marks of student 4 is ",total4)


```

```python
student_averages = marks.mean(axis=1)
print("Student averages:", student_averages)

students_above_80_avg = np.where(student_averages > 80)[0]

if len(students_above_80_avg) > 0:
    print("Students with average marks above 80:")
    for student_index in students_above_80_avg:
        print(f"  Student {student_index + 1} (Average: {student_averages[student_index]:.2f})")
else:
    print("No students found with average marks above 80.")
```

Normalize this array to the range 0–1:
arr = np.array([10, 20, 30, 40, 50])

Use the min-max normalization formula. Don’t hardcode the minimum or maximum.

```python
arr = np.array([10, 20, 30, 40, 50])
min_val=arr.min()
max_value=arr.max()
normalization = (arr-min_val)/(max_value-min_val)
print("Normalized array ", normalization)
```

marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])

Find all of these:

Total marks of every student using one line

Average marks of every subject using one line

Index of the student with the highest total

Index of the subject with the highest average

Print the actual marks row of the best student


```python
marks = np.array([
    [78, 85, 90],
    [65, 70, 72],
    [92, 88, 95],
    [55, 60, 58]
])

total_marks=marks.sum(axis=1)
print("total marks of each stuent",total_marks)

avg_marks=marks.mean(axis=0)
print("Average marks of each subject",avg_marks)

index_high=total_marks.argmax()
print("index of student with high marks",index_high)

index_avg=avg_marks.argmax()
print("index of subject with highest average",index_avg)

best_student=marks[index_high]
print("marks of best student",best_student)
```

salaries = np.array([
  
    25000, 40000, 55000,
    70000, 90000, 120000
])

Apply these bonuses:

salary < 40000        → 10% bonus

40000 to 70000        → 7% bonus

salary > 70000        → 5% bonus



```python
salaries = np.array([
    25000, 40000, 55000,
    70000, 90000, 120000
])

# Apply 10% bonus for salary < 40000
salaries[salaries<40000] = salaries[salaries<40000] * 1.1

# Apply 7% bonus for 40000 <= salary <= 70000
salaries[(salaries>=40000) & (salaries<=70000)] = salaries[(salaries>=40000) & (salaries<=70000)] * 1.07

# Apply 5% bonus for salary > 70000
salaries[salaries>70000] = salaries[salaries>70000] * 1.05

salaries
```
