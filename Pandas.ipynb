```python
#finding how much better the ratings are compared to the average
review_rating_mean=review.rating.mean()
review.rating.map(lambda p : p - review_rating_mean)
```

```python
#using apply()
def remean_row(row):
  row.rating=row.rating-review_rating_mean
  return row
review.apply(remean_row,axis='columns')
```
